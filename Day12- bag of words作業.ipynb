## 作業目標：搭建一個bag of words模型

---

```python
import pandas as pd
import nltk

import numpy as np
dataset=pd.read_csv('Restaurant_Reviews.tsv',delimiter='\t',quoting=3)
corpus=dataset['Review'].values
```

```python
nltk.download('punkt')
```

```python
corpus[0:10]
```

### 從文本中取出所有單字

```python
import re
```

```python
whole_words = []
for sentence in corpus:
    sentence = re.sub(r'[^\w ]','',sentence.lower())
    tokenized_sentence = nltk.tokenize.word_tokenize(sentence)
    for word in tokenized_sentence:
        whole_words.append(word)
```

```python
whole_words = set(whole_words)
```

### 移除重複單字

```python

print('共有{}個單字'.format(len(whole_words)))
```

### 建立字典使每一個單字有對應數值

```python
word_index = {}
index_word = {}
n = 0
for i, word in enumerate(whole_words):
    """自行填入"""
    word_index[word] = i
    index_word[i] = word
    
```

## 轉換句子為bag of words型式

```python
np.zeros(len(word_index.keys()))
```

```python
len(word_index.keys())
```

```python
def _get_bag_of_words_vector(sentence, word_index_dic, whole_words):
    sentence = sentence
    '''創建一個vector'''
    vector =np.zeros(len(word_index.keys()))
    for word in nltk.word_tokenize(sentence):
        if word in whole_words:
            '''自行填入'''
            vector[word_index_dic[word]] = 1
    return vector
```

```python
print(_get_bag_of_words_vector('Wow... Loved this place.', word_index, whole_words))
```

```python
sum(_get_bag_of_words_vector('Wow... Loved this place.', word_index, whole_words))
```
